```python
#Lets distribute cabin into bins on the basis of people survived and we will fill the null values accordingly
```

```python
df['FamilySize']=df['SibSp']+df['Parch']+1
```

```python
df.drop(columns=['SibSp','Parch','Age'],inplace=True)
```

```python
df
```

```python
df['Cabin'].value_counts()
```

```python
df['Cabin']=df['Cabin'].fillna('z')
```

```python
df
```

```python
df['Cabin_category']=df['Cabin'].apply(lambda a:a[0])
```

```python
df.drop(columns='Cabin',inplace=True)
```

```python
df = pd.get_dummies(df, columns=['Cabin_category'], drop_first=True)
```

```python
df
```

```python
df['Embarked']=df['Embarked'].fillna('N')
```

```python
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
```

```python
df
```

```python
# Prefix
def get_ticket_prefix(ticket):
    ticket = str(ticket)
    if ticket.isdigit():
        return 'NUM'
    return ticket.split()[0]

df['Ticket_prefix'] = df['Ticket'].apply(get_ticket_prefix)

# Frequency
df['Ticket_count'] = df.groupby('Ticket')['Ticket'].transform('count')

# One-hot encoding
df = pd.get_dummies(df, columns=['Ticket_prefix'], prefix='Ticket')
```

```python
df['Title'] = df['Name'].str.extract('([A-Za-z]+)\.', expand=False)
```

```python
df['Title'] = df['Title'].replace([
    'Capt', 'Col', 'Major', 'Dr', 'Rev'
], 'Officer')

df['Title'] = df['Title'].replace([
    'Lady', 'Countess', 'Sir', 'Don', 'Jonkheer', 'Dona'
], 'Royalty')
```

```python
df = pd.get_dummies(df, columns=['Title'], prefix='Title')
```

```python
df
```

```python
df.drop(columns=['Ticket','Name'],inplace=True)
```

```python
df
```

```python
df['Sex']=df['Sex'].map({'male':0,'female':1})
```

```python
df
```

```python
df=df.astype(int)
```

```python
df
```

```python
df.drop(columns=['PassengerId'], inplace=True)
```
